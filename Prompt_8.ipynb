```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
```

```python
def sin(x):
    return np.sin(x)

def cos(x):
    return np.cos(x)
```

```python
fig, axes = plt.subplots(1,2)

axes[0].plot(x,sin(x))
axes[1].plot(x,cos(x))

axes[0].set_xlabel("Sin Curve")
axes[1].set_xlabel("Cos Curve")

plt.savefig("prompt8.pdf")
plt.show()
```
